```python
from sklearn.datasets import load_iris
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier

# 데이터 로드
iris = load_iris()
X = iris.data
y = iris.target
model = RandomForestClassifier()
kf= KFold(n_splits=5, shuffle=True)
scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
print("교차 검증 정확도 : ", scores)
print("평균 정확도 : ", scores.mean())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 3.5))

plt.subplot(121)
plt.plot(X[y==0, 1], X[y==0, 3], "yo", label="Iris setosa")
plt.plot(X[y==1, 1], X[y==1, 3], "bs", label="Iris versicolor")
plt.plot(X[y==2, 1], X[y==2, 3], "g^", label="Iris virginica")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.grid()
plt.legend()

plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c="k", marker=".")
plt.xlabel("Petal length")
plt.tick_params(labelleft=False)
plt.gca().set_axisbelow(True)
plt.grid()

plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0, n_init="auto").fit(X[:, 2:])
kmeans.labels_
kmeans.cluster_centers_
```
